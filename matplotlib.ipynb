### Importing Matplotlib

```python
import matplotlib.pyplot as plt
```

### linear Plot

```python
x = [60, 30, 45, 100]
y = [10, 5, 11, 7]
plt.plot(x, y, marker = "o")
# separate graph at x- axis
plt.plot(x, marker = "^")
# separate graph at y- axis
plt.plot(y, marker = "^")
plt.show()
```
