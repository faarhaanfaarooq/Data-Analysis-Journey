### Importing Matplotlib

```python
import matplotlib.pyplot as plt
```

### linear Plot

```python
x = [60, 30, 45, 100]
y = [10, 5, 11, 7]
plt.plot(x, y, marker = "o")
# separate graph at x- axis
plt.plot(x, marker = "^")
# separate graph at y- axis
plt.plot(y, marker = "^")
plt.show()
```

### Bar Graph

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
plt.bar(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
c = ['r', 'g', 'b', 'y', 'b' ]
plt.bar(x, y, color = c)
plt.show()
```

### showing bar graph horizontally

```python
import numpy as np
```

```python
x = np.array(['a', 'b', 'c', 'd'])
y = np.array([10, 20, 30, 40])
plt.barh(x, y)
plt.show()
```

```python
x = np.array(['a', 'b', 'c', 'd'])
y = np.array([10, 20, 30, 40])
plt.bar(x, y, width=0.1)
plt.show()
```

### ScatterPlot

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
plt.title("Scatter Plot")
plt.xlabel("Person")
plt.ylabel("Numbers")
plt.scatter(x, y)
plt.show()
```
