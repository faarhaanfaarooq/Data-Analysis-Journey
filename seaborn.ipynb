```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Matplotlib

```python
var1 = [10, 20, 30, 40, 50]
var2 = [60, 70, 80, 90, 100]
```

```python
plt.plot(var1, var2)
plt.show()
```

### Seaborn

```python
var1 = [10, 20, 30, 40, 50]
var2 = [60, 70, 80, 90, 100]
```

```python
sns.lineplot(x = var1, y = var2)
plt.show()
```

### Graph from dataframe

```python
var1 = [10, 20, 30, 40, 50]
var2 = [60, 70, 80, 90, 100]

df = pd.DataFrame({"X":var1, "Y":var2})

sns.lineplot(df)
plt.show()
```

```python
pendf = sns.load_dataset("penguins")
```

```python
pendf
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = pendf)
plt.show()
```

### Building the same graph for categorical division

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = pendf, hue = "sex")
plt.show()
```

### Bar Graph

```python
dfOne = sns.load_dataset("penguins")
dfOne
```

```python
sns.barplot(x = dfOne.island, y = dfOne.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = dfOne, hue = "sex")
plt.show()
```

```python
islandOrder = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = dfOne, hue = "sex", order = islandOrder)
plt.show()
```

### Histogram

```python
sns.displot(dfOne["bill_length_mm"])
plt.show()
```

```python
sns.displot(dfOne["bill_length_mm"], bins = [30, 40, 50, 60])
plt.show()
```

```python
sns.displot(dfOne["bill_length_mm"], bins = [30, 40, 50, 60], kde = True)
plt.show()
```
