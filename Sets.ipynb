# Characteristics of Sets:

  **Unordered**: The elements in a set do not have a defined order. You cannot refer to an item by index.

  **Mutable**: You can add or remove items from a set.

  **No Duplicates**: Sets cannot have two items with the same value.

### When to Use Sets:

  * When you need to ensure elements are unique.
  * For efficient membership testing (checking if an element is in a collection).
  * When you need to perform set operations like unions and intersections.


---



### Creating a set

```python
my_set = {1,2,3,4,5}

#Creating a set with constructor
new_set = set([1,2,3,4,5,5,6])

print(my_set)
print(new_set)
```

### Adding elements

```python
my_set.add(9)
print(my_set)
```

### Removing Elements

```python
#Raises an error if not present
my_set.remove(9)
print(my_set)

#Removes from the front and returns an arbitrary set element
my_set.pop()
print(my_set)
```

# Operations on Sets:

Python sets support a variety of operations that correspond to mathematical set operations.

  **Union ('|')**: Combines two sets to form a new one containing items in either.

  **Intersection ('&')**: Gets items only in both.

  **Difference ('-')**: Gets items in the first set but not in the second.

  **Symmetric Difference ('^')**: Gets items in either set, but not in both.

---

### Manually Operated Operations

```python
setA = {1,2,3,4,5}
setB = {4,5,6,7,8}

#Union
print(setA | setB)

#Intersection
print(setA & setB)

#Difference
print(setA - setB)

#Symmetric Difference
print(setA ^ setB)
```
